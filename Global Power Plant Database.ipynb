```python
#Dropping irrelavent columns
df.drop(columns={'estimated_generation_gwh','year_of_capacity_data','wepp_id','other_fuel1','other_fuel2','other_fuel3','name','gppd_idnr','url','country_long'},axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.skew()
```

Now we will fill the the categorical column null values with mode of the column and numerical column null values with mean or median of the column.

```python
df['capacity_mw']=df['capacity_mw'].fillna(df['capacity_mw'].mean())
df['latitude']=df['latitude'].fillna(df['latitude'].mean())
df['longitude']=df['longitude'].fillna(df['longitude'].mean())
df['commissioning_year']=df['commissioning_year'].fillna(df['commissioning_year'].mean())
```

```python
df['generation_gwh_2013']=df['generation_gwh_2013'].fillna(df['generation_gwh_2013'].median())
df['generation_gwh_2014']=df['generation_gwh_2014'].fillna(df['generation_gwh_2014'].median())
df['generation_gwh_2015']=df['generation_gwh_2015'].fillna(df['generation_gwh_2015'].median())
df['generation_gwh_2016']=df['generation_gwh_2016'].fillna(df['generation_gwh_2016'].median())
df['generation_gwh_2017']=df['generation_gwh_2017'].fillna(df['generation_gwh_2017'].median())
df['generation_gwh_2018']=df['generation_gwh_2018'].fillna(df['generation_gwh_2018'].median())
```

```python
df['primary_fuel']=df['primary_fuel'].fillna(df['primary_fuel'].mode()[0])
df['owner']=df['owner'].fillna(df['owner'].mode()[0])
df['geolocation_source']=df['geolocation_source'].fillna(df['geolocation_source'].mode()[0])
df['generation_data_source']=df['generation_data_source'].fillna(df['generation_data_source'].mode()[0])
```

```python
#Data type of generation_gwh_2019 is object we need to change it to float then fill the null values
df['generation_gwh_2019'] = pd.to_numeric(df['generation_gwh_2019'],errors='coerce')
print(df['generation_gwh_2019'].dtypes)
```

```python
df['generation_gwh_2019']=df['generation_gwh_2019'].fillna(df['generation_gwh_2019'].median())
```

```python
df.isnull().sum()
```

```python
#Extract power plant age from commissioning year and drop this year column
df['Power_plant_age']= 2019- df['commissioning_year']
df.drop(columns=['commissioning_year'], inplace=True)
```

# #Prediction for capacity_mw

```python
#separating target column and columns not necessary in the analysis from data
X= df.drop(['capacity_mw'],axis=1)
Y= df['capacity_mw']
```

```python
print(X)
print(Y)
```

# Splitting the data into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

# Creating model

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
model= LinearRegression()
model.fit(X_train, Y_train)
```

```python
#prediction on training data
prediction= model.predict(X_train)
r2_train= metrics.r2_score(Y_train,prediction)
print('R squared value:',r2_train)
```

```python
#prediction on testing data
prediction1= model.predict(X_test)
r2_test= metrics.r2_score(Y_test,prediction1)
print('R squared value:',r2_test)
```
